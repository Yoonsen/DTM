```python
counts = dh.Counts(c2)
```

```python
tot = dh.totals(500000)
```

```python
df = counts.frame
```

```python
df.sum(axis = 1)
```

```python
combo = pd.DataFrame(df.sum(axis=1), columns=['freq'])
```

```python
combo["tot"] = tot.freq
```

```python
combo.tot = combo.tot.fillna(combo.tot.min()/2)
```

```python
combo['diff'] = (combo.freq/combo.freq.sum())*100/(combo.tot/combo.tot.sum())
```

```python
docfreq = (df/df).fillna(0)
docfreq
```

```python
ds = docfreq.sum(axis = 1)
```

```python
# Ensure that the index of combo is a column if it's not already
combo.reset_index(inplace=True)

# Rename the column to ensure clarity (assuming the column with words in combo is named 'word')
combo.rename(columns={'index': 'word'}, inplace=True)

# Map the document frequencies from docfreq to the new column in combo
combo['docfreq'] = combo['word'].map(lambda x:ds[x])

# If you want to set 'word' back as index
combo.set_index('word', inplace=True)
```

```python
combo.to_csv("combo_andre_del.csv")
```
